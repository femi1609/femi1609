The pytube library allows you to work with YouTube videos and provides functionalities to download or extract information from YouTube videos.

1. Defining the function

2. Extracting youtube video data

3. Getting video title and views

4. Printing the extracted data

5. Handling exceptions & defining the main function

```python
from pytube import YouTube

def extract_youtube_data(video_url):
    try:

        yt = YouTube(video_url)
        video_title = yt.title
        video_views = yt.views
        print("\nTitle:", video_title)
        print("\nViews:", video_views)

    except Exception as e:
        print("Error:", str(e))
def main():
    video_url = input("Enter the youtube video URL: ")
    extract_youtube_data(video_url)
```

```python
main()
```

###**ANALYSIS OF YOUTUBE VIDEO COMMENTS BASED ON WORD COUNT(DESC).**

1. Install the required library
   
2. Import the required modules

3. Define the count_words function

4. Initialize the YoutubeCommentDownloader

5. Fetch and sort comments

6. Sort the comments by word count

7. Print the top 10 comments


```python
from itertools import islice
from youtube_comment_downloader import *

def count_words(comment):
    return len(comment['text'].split())

downloader = YoutubeCommentDownloader()
video_url = input("Enter the YouTube video URL: ")
comments = downloader.get_comments_from_url(video_url, sort_by=SORT_BY_POPULAR)

# Sort the comments by the number of words
comments_sorted_by_words = sorted(comments, key=lambda x: count_words(x), reverse=True)

for comment in islice(comments_sorted_by_words, 10):
    print(f"Words: {count_words(comment)}, Comment: {comment['text']}\n")
```

###**COMMENT ANALYSIS AND SORTING BASED ON WORD COUNT(ASC)**

1. Install the required libraries

2. Import the required modules

3. Define the count_words function

4. Initialize the YoutubeCommentDownloader

5. Fetch and store comments in a DataFrame

6. Create a DataFrame from the comment data

7. Sort the DataFrame by word count- Sort the DataFrame by the 'Words' column in ascending order:

8. Print the top 10 comments



```python
import pandas as pd
from itertools import islice
from youtube_comment_downloader import *

def count_words(comment):
    return len(comment['text'].split())

downloader = YoutubeCommentDownloader()
video_url = input("Enter the YouTube video URL: ")
comments = downloader.get_comments_from_url(video_url, sort_by=SORT_BY_POPULAR)

comment_data = [{'Words': count_words(comment), 'Comment': comment['text']} for comment in comments]

df = pd.DataFrame(comment_data)
df_sorted = df.sort_values(by='Words')

for _, row in islice(df_sorted.iterrows(), 80):
    print(f"Words: {row['Words']}, Comment: {row['Comment']}\n")
df
```

###**VISUALISATION USING WORD CLOUD**

1. Import necessary libraries:
  
   * WordCloud: To create the word cloud.
   * STOPWORDS: A set of common stopwords to filter out from the comments.
   * matplotlib.pyplot: To display the word cloud.
2. Initialize variables:
   - comment_words: An empty string that will store all the words from the comments.
   - stopwords: A set of common stopwords to filter out from the comments.

3. Iterate through the comments in your DataFrame (df.Comment):
   - Convert each comment to a string.
   - Split the comment into individual words.
   - Convert each word to lowercase and join them back into a single string.
4. Create the WordCloud object:
   - Set the desired parameters, such as width, height, background color, stopwords, and minimum font size.
   - Generate the word cloud from the comment_words string.

5. Plot and display the WordCloud image

```python
pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
comment_words = ''
stopwords = set(STOPWORDS)

for val in df.Comment:

    val = str(val)

    tokens = val.split()

    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

###**EXTRACTING VIDEO DETAILS USING URL**
1. Importing the module & extracting video title

2. Extracting views count

3. Extracting video description

4. Extracting video tags

5. Extracting video date published & upload Date

6. Extracting video genre

7. Extracting channel name & URL

8. Extracting channel subscribers count



```python
import re
import json

link = "https://www.youtube.com/live/VNvU37jqiWM?si=ZTZovKCQDji145GE"
response = requests.get(link)
soup = BeautifulSoup(response.text, "html.parser")

titleSoupMeta = soup.find("meta", property="og:title")
videoTitle = titleSoupMeta["content"] if titleSoupMeta else "NotFound"
print(f'\nVideo Title:{videoTitle}')

thumbSoupMeta = soup.find("meta", property="og:image")
videoImage = thumbSoupMeta["content"] if thumbSoupMeta else "NotFound"
print(f'\nVideo Image:{videoImage}')

views = soup.find("meta", itemprop="interactionCount")['content']
print(f'\nViews:{views}')

description = soup.find("meta", itemprop="description")['content']
print(f"\nVideo Description:{description}")


tags = ', '.join([ meta.attrs.get("content") for meta in soup.find_all("meta", {"property": "og:video:tag"}) ])
print(f"\nTags:{tags}")

date_published = soup.find("meta", itemprop="datePublished")['content']
print(f"\nVideo Date Published:{date_published}")

uploadDate = soup.find("meta", itemprop="uploadDate")["content"]
print(f'\nVideo Upload Date:{uploadDate}')

genre = soup.find("meta", itemprop="genre")["content"]
print(f'\nGenre:{genre}')

data = re.search(r"var ytInitialData = ({.*?});", soup.prettify()).group(1)
data_json = json.loads(data)
videoPrimaryInfoRenderer = data_json['contents']['twoColumnWatchNextResults']['results']['results']['contents'][0]['videoPrimaryInfoRenderer']
videoSecondaryInfoRenderer = data_json['contents']['twoColumnWatchNextResults']['results']['results']['contents'][1]['videoSecondaryInfoRenderer']

channel_name = soup.find("span", itemprop="author").next.next['content']
print(f'\nChannel Name:{channel_name}')
channel_url = soup.find("span", itemprop="author").next['href']
print(f'\nChannel URL:{channel_url}')

channel_subscribers = videoSecondaryInfoRenderer['owner']['videoOwnerRenderer']['subscriberCountText']['accessibility']['accessibilityData']['label']
print(f'\nChannel Subscribers:{channel_subscribers}')
```
